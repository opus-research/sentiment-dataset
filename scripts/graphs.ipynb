```python
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = ['100% (Full)', '66% (Partial)', '0% (None)']
part1_values = [3.85, 3.37, 3.09]
part2_values = [4.19, 3.72, 3.56]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, part1_values, width, label='First Pass')
bars2 = ax.bar(x + width/2, part2_values, width, label='Second Pass')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Agreement Level')
ax.set_ylabel('Confidence Levels (1-5)')
ax.set_yticks([1,2,3,4,5])
ax.set(ylim=[0,5])
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

ax.bar_label(bars1, padding=3)
ax.bar_label(bars2, padding=3)

fig.tight_layout()

plt.savefig("confidence.pdf")
```

```python
import matplotlib.pyplot as plt

# Original dictionary of emotions and their frequencies
emotions = {'Joy': 1770,
            'Anger': 461,
            'Fear': 181,
            'Sadness': 1121,
            'Surprise (Negative)': 307,
            'Surprise (Positive)': 80,
            'Love': 153}

# Assign a color based on the emotion type
emotions_colored = {emotion: {'freq': freq, 'color': 'mediumseagreen' if emotion in ['Joy', 'Love', 'Surprise (Positive)'] else 'indianred'}
                    for emotion, freq in emotions.items()}

# Sort emotions based on color and then frequency
sorted_emotions = sorted(emotions_colored.items(), key=lambda x: (x[1]['color'], -x[1]['freq']))

# Extracting sorted labels, frequencies, and colors
labels = [emotion[0] for emotion in sorted_emotions]
frequencies = [emotion[1]['freq'] for emotion in sorted_emotions]
colors = [emotion[1]['color'] for emotion in sorted_emotions]

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(labels, frequencies, color=colors)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')

plt.ylabel('Frequency (2nd pass)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("emofrequency.pdf")
```
